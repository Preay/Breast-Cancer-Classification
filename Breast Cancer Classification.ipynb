## CASE STUDY: BREAST CANCER CLASSIFICATION

 
    

### Problem statement:
1.Predicting if the cancer diagnosis is benign or malignant based on several observations/features.
2. 30 features are used
3. total number of instances:569

### step 1 :Importing datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
cancer
```

```python
df = pd.DataFrame(np.c_[cancer["data"],cancer["target"]],columns=np.append(cancer["feature_names"],["target"]))
df
```

```python
df.info()
```

## Visualizing Data

```python
sns.pairplot(df,hue='target' , vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area',
        'mean smoothness', 'mean compactness', 'mean concavity'])

#blue is milignet case(bad one)
#orange is good
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
```

## Model Training

```python
x= df.drop(["target"],axis=1)
y= df["target"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.2,random_state=42)
x_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix

```

```python
model=SVC()
model.fit(x_train,y_train)
```

## Model Evaluation

```python
y_pred=model.predict(x_test)
y_pred
```

```python
confusion_matrix(y_test,y_pred)
```

```python
con_mat= confusion_matrix(y_test,y_pred)
sns.heatmap(con_mat,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

## conclusion

svm was able to classify tumors Malignant/Benign with accuracy of 92%
